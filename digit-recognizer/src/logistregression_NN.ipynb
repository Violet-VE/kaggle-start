## 2.design NN model use logist regression

### 2.1 sigmod function

```python
def sigmod(x):
    s = 1/(1+np.exp(-x))
    
    return s
```

```python
print(sigmod(2))
```

### 2.2 cost function

```python
def cost(yhat, y):
    m = y.shape[0]
    
    loss = np.dot(y, np.log(yhat)) + np.dot((1-y), np.log(1-yhat))
    cost = (-1/m) * np.sum(loss)
    
    return cost
```

```python
yhat = np.array([0.3,0.2])
y = np.array([1,0])

print(cost(yhat, y))
```
